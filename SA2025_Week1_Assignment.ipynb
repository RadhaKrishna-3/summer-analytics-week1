### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
avg_mpg = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.lineplot(data=avg_mpg, x='model_year', y='mpg', hue='origin', marker='o',
             palette='tab10', style='origin', dashes=False)

plt.title('Evolution of Average MPG Over Years by Origin')
plt.xlabel('Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.tight_layout()
plt.show()
```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
import pandas as pd

# Step 1: Group by all columns except mpg and model_year
group_cols = ['cylinders', 'displacement', 'horsepower', 'weight',
              'acceleration', 'origin']

# Create a synthetic "model_id"
df['model_id'] = df[group_cols].astype(str).agg('-'.join, axis=1)

# Step 2: For each model_id, count unique model_years
year_counts = df.groupby('model_id')['model_year'].nunique()

# Step 3: Keep those model_ids appearing in more than one year
multi_year_models = year_counts[year_counts > 1].index

# Step 4: Compute std and mean mpg for those models
stats = (
    df[df['model_id'].isin(multi_year_models)]
    .groupby('model_id')
    .agg(
        num_appearances=('model_year', 'count'),
        mpg_std=('mpg', 'std'),
        avg_mpg=('mpg', 'mean')
    )
    .reset_index()
)

# Step 5: Filter as requested
result = stats[(stats['mpg_std'] < 1.0) & (stats['num_appearances'] >= 2)]

# Step 6: Find minimum std
min_std = result['mpg_std'].min()
best = result[result['mpg_std'] == min_std]

# Step 7: Sort as bonus
best_sorted = best.sort_values(['num_appearances', 'avg_mpg'], ascending=[False, False])

# Show relevant columns
print(best_sorted[['model_id', 'num_appearances', 'avg_mpg']])
```
